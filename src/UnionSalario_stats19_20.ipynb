```python
import pandas as pd
```

```python
salario = pd.read_csv("../NBA_csv/salario19_20.csv")

stats = pd.read_csv("../NBA_csv/Stats19_20.csv")
                      
```

```python
stats
```

```python
salario.rename(columns = {"Jugador":"Player"}, inplace = True)
```

```python
stats.drop(columns = ["Rk","Player-additional"], inplace =True)
```

```python
stats.Pos.unique()
diccionario_posiciones = {
    'PG': 'BA',
    'SG': 'ES',
    'SF': 'AL',
    'PF': 'AP',
    'C': 'P',
    "PF-C":"P",
    'SG-PG':"ES",
    'SG-SF':"AL",
    'C-PF':"AP",
    'PG-SG':"BA",
    'SF-SG':"ES",
    'PF-SF':"AP",
    'SF-PF':"AP",
}
stats['Pos'] = stats['Pos'].map(diccionario_posiciones)
```

```python
salario=salario[salario["Salario"]!= 0 ]
salario
```

```python
statsALL20_21 =pd.merge(salario, stats, on='Player', how='inner')
```

```python
statsALL20_21=statsALL20_21.dropna()
```

```python
statsALL20_21.to_csv("statsALL19_20.csv", index= False)
```
