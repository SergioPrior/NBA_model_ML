```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 10)
```

```python
 # conferencia_east
# Crear una visualizacin con seaborn y matplotlib
plt.figure(figsize=(10, 6))
sns.barplot(x='Pg', y='nombre-equipo', data=conferencia_east, palette='viridis')
plt.title('Clasificacin de Equipos')
plt.xlabel('Puntos')
plt.ylabel('Nombre del Equipo')
plt.show()
```
