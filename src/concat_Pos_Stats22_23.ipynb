```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

```

```python
stats = pd.read_csv("../NBA_csv/stats_player_22_23.csv")

position = pd.read_csv("../NBA_csv/posicions_player.csv")
```

```python
stats = stats.drop(columns="Rk")

position=position.drop(columns = ["No.","Ht","Wt","Birth Date","Unnamed: 6","Exp","College","Player-additional"])

position = position.drop_duplicates()
```

```python
statsposition = pd.merge(stats, position, on='Player', how='inner')
```

```python
diccionario_posiciones = {
    'PG': 'BA',
    'SG': 'ES',
    'SF': 'AL',
    'PF': 'AP',
    'C': 'P'
}
statsposition['Pos'] = statsposition['Pos'].map(diccionario_posiciones)
```

```python
statsposition=statsposition.drop_duplicates(subset="Player")
```

```python
dupli = statsposition[statsposition['Player'].duplicated()]
```

```python
statsposition= statsposition.dropna()
```

```python
statsposition = statsposition.reset_index()
statsposition['Player'] = statsposition['Player'].str.capitalize()
```

```python
statsposition = statsposition.drop(columns = "index")
```

```python
statsposition.to_csv("stats22_23_clean.csv", index = False)
```
