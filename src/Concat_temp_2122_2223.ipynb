```python
import pandas as pd

pd.options.mode.copy_on_write = True
```

```python
salario21_22  = pd.read_excel("../NBA_csv/salario21_22.xlsx")

salario22_23 = pd.read_csv("../NBA_csv/modelo22_23.csv")
```

```python
#Eliminamos columnas que no queremos.
salario21_22.drop(columns=["JUGADOR","Rk","Tm"], inplace = True)
```

```python
salario21_22.dropna(inplace = True)
```

```python
# Pasamos posiciones al castellano.

diccionario_posiciones = {
    'PG': 'BA',
    'SG': 'ES',
    'SF': 'AL',
    'PF': 'AP',
    'C': 'P'
}

salario21_22["Pos"] = salario21_22["Pos"].map(diccionario_posiciones)
```

```python
#Eliminamos de la columna salarios el simbolo de dolar y los puntos para poder usarla como dato numerico.
salario21_22["SALARIO 2021-22"] = salario21_22["SALARIO 2021-22"].str.replace("$","").str.replace(".","")

salario21_22.rename(columns= {"SALARIO 2021-22":"Salario"}, inplace = True)
```

```python
salario21_22
```

```python
#concatenamos las dos temporadas para tener un modelo mas eficiente
total_temporadas_train = pd.concat([salario21_22, salario22_23], ignore_index = True )
```

```python
# pasamos a mayuscula solo la primera palabra para tener un nivel de busqueda mas optimo
total_temporadas_train['Player'] = total_temporadas_train['Player'].apply(lambda x: x.capitalize())
```

```python
total_temporadas_train = total_temporadas_train[~total_temporadas_train['Salario'].str.contains('Dual',
                                                                                                case=False, na=False)]
```

```python
#Pasamos el valor a entero para trabajar con el.
total_temporadas_train.Salario = total_temporadas_train.Salario.astype(int)
```

```python
#convertimos a csv.
total_temporadas_train.to_csv("total_temporadas_train.csv", index = False)
```
